```python
# Import required libraries
import pandas as pd  # For data manipulation
from sklearn.model_selection import train_test_split, GridSearchCV  # For data splitting and hyperparameter tuning
from sklearn.preprocessing import StandardScaler  # For feature scaling
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay  # For model evaluation
import xgboost as xgb  # For XGBoost classification
import matplotlib.pyplot as plt  # For data visualization
import seaborn as sns  # For enhanced visualizations
import joblib  # For saving models

# Define CSV file paths for all companies
csv_files = {
    "TATASTEEL.NS": "E:/m0161finale/m0161final/m0161final/tatasteel_ns_3yr_data.csv",  # Tata Steel stock data
    "TCS.NS": "E:/m0161finale/m0161final/m0161final/TCS_data.csv",  # TCS stock data
    "HDFCBANK.NS": "E:/m0161finale/m0161final/m0161final/HDFCBank_3yr_data.csv",  # HDFC Bank stock data
    "INFY.NS": "E:/m0161finale/m0161final/m0161final/Infosys_data.csv",  # Infosys stock data
    "WIPRO.NS": "E:/m0161finale/m0161final/m0161final/Wipro_data.csv"  # Wipro stock data
}

# Loop through each company's data file
for company, csv_file_path in csv_files.items():
    print(f"Processing data for {company}...")

    # Load the stock market data from CSV file
    data = pd.read_csv(csv_file_path)

    # Feature engineering: calculate previous close price
    data['Prev_Close'] = data['Close'].shift(1)
    
    # Calculate price change
    data['Price_Change'] = data['Close'] - data['Prev_Close']
    
    # Calculate percentage change in price
    data['Percent_Change'] = data['Price_Change'] / data['Prev_Close'] * 100
    
    # Create target column: 1 if price increased, 0 otherwise
    data['Target'] = (data['Price_Change'] > 0).astype(int)
    
    # Drop rows with NaN values (first row will be NaN due to shifting)
    data = data.dropna()

    # Create visualization subplots
    plt.style.use('classic')  # Using classic style instead of seaborn
    fig = plt.figure(figsize=(20, 15))
    
    # 1. Price Trends with Moving Averages
    ax1 = plt.subplot(3, 2, 1)
    data['MA20'] = data['Close'].rolling(window=20).mean()
    data['MA50'] = data['Close'].rolling(window=50).mean()
    ax1.plot(data.index, data['Close'], label='Close Price', linewidth=1)
    ax1.plot(data.index, data['MA20'], label='20-day MA', linewidth=1.5)
    ax1.plot(data.index, data['MA50'], label='50-day MA', linewidth=1.5)
    ax1.set_title(f'{company} Price Trends with Moving Averages', pad=10, fontsize=12)
    ax1.set_xlabel('Time', fontsize=10)
    ax1.set_ylabel('Price', fontsize=10)
    ax1.legend(fontsize=8)
    ax1.grid(True, alpha=0.3)
    
    # 2. Volume Analysis
    ax2 = plt.subplot(3, 2, 2)
    ax2.bar(data.index, data['Volume'], alpha=0.7, color='blue')
    ax2.set_title(f'{company} Trading Volume', pad=10, fontsize=12)
    ax2.set_xlabel('Time', fontsize=10)
    ax2.set_ylabel('Volume', fontsize=10)
    ax2.grid(True, alpha=0.3)
    
    # 3. Daily Returns Distribution
    ax3 = plt.subplot(3, 2, 3)
    sns.histplot(data=data, x='Percent_Change', bins=50, ax=ax3, color='skyblue')
    ax3.set_title(f'{company} Daily Returns Distribution', pad=10, fontsize=12)
    ax3.set_xlabel('Daily Returns (%)', fontsize=10)
    ax3.set_ylabel('Frequency', fontsize=10)
    ax3.grid(True, alpha=0.3)
    
    # 4. Price vs Volume Scatter
    ax4 = plt.subplot(3, 2, 4)
    ax4.scatter(data['Volume'], data['Close'], alpha=0.5, color='green')
    ax4.set_title(f'{company} Price vs Volume', pad=10, fontsize=12)
    ax4.set_xlabel('Volume', fontsize=10)
    ax4.set_ylabel('Price', fontsize=10)
    ax4.grid(True, alpha=0.3)
    
    # 5. Volatility (Rolling Standard Deviation)
    ax5 = plt.subplot(3, 2, 5)
    data['Volatility'] = data['Percent_Change'].rolling(window=20).std()
    ax5.plot(data.index, data['Volatility'], color='red', linewidth=1)
    ax5.set_title(f'{company} 20-day Rolling Volatility', pad=10, fontsize=12)
    ax5.set_xlabel('Time', fontsize=10)
    ax5.set_ylabel('Volatility', fontsize=10)
    ax5.grid(True, alpha=0.3)
    
    plt.tight_layout(pad=3.0)
    plt.show()

    # Define features for classification
    features = ['Prev_Close', 'Volume', 'Percent_Change']
    X = data[features]
    y = data['Target']

    # Standardize features for better model performance
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Split dataset into training (80%) and testing (20%)
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

    # Define hyperparameter grid for XGBoost tuning
    param_grid = {
        'max_depth': [3, 5, 7],  # Maximum tree depth
        'learning_rate': [0.01, 0.1, 0.3],  # Learning rate
        'n_estimators': [100, 200],  # Number of boosting rounds
        'subsample': [0.8, 1.0],  # Subsample ratio of training instances
        'colsample_bytree': [0.8, 1.0]  # Subsample ratio of columns when constructing each tree
    }
    
    # Create XGBoost classifier
    xgb_model = xgb.XGBClassifier(objective='binary:logistic', random_state=42, use_label_encoder=False)
    
    # Perform grid search cross-validation for XGBoost
    grid_search = GridSearchCV(xgb_model, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)

    # Retrieve the best model from grid search
    best_model = grid_search.best_estimator_

    # Make predictions on test data
    y_pred = best_model.predict(X_test)

    # Evaluate model performance using confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Down", "Up"])
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f"Confusion Matrix for {company}")
    plt.show()

    # Print classification report (precision, recall, f1-score)
    print(f"Classification Report for {company}:")
    print(classification_report(y_test, y_pred))

    # Save trained XGBoost model and scaler for future use
    model_filename = f"E:/m0161finale/m0161final/m0161final/xgb_{company}_classifier.pkl"
    scaler_filename = f"E:/m0161finale/m0161final/m0161final/scaler_{company}.pkl"
    joblib.dump(best_model, model_filename)
    joblib.dump(scaler, scaler_filename)
    print(f"XGBoost Classifier model for {company} saved as {model_filename}.")
    print(f"Scaler for {company} saved as {scaler_filename}.")

    print(f"Completed processing for {company}.\n{'-'*60}")
```
